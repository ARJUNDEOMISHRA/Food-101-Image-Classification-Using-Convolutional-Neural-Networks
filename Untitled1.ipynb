# Create preprocessing functions for our data

Neural network perform best when data is in a certain way (eg. batched,normalized,etc)
however, not all data(including data from tensorflow dataset)comes like this.
so in order to get it ready for a neural network ,you'll often have to write preprocessing functions and map it to your data.
what  we know about our data :
in uint8 dataype
comprised of all different sizes tensors(different sized images)

what we know models like :
Data in float 32 dtype (or for mixed precision float16 & float 32)
for batches ,tensorflow likes all of the tensors within a batch to be of the same size 
scaled (values between 0&1) also called normalized tensors generally perform better

with these points in mind , we 've got a few things we can tackle with a preprocessing function.

```python
def preprocess_img(image, label, img_shape=224):
    """
    Converts image datatype from 'uint8'  'float32' and reshapes
    image to [img_shape, img_shape, color_channels]
    """
    image = tf.image.resize(image, [img_shape, img_shape])
    return tf.cast(image, tf.float32), label
```

```python
# Preprocess a single sample image
preprocessed_img, preprocessed_label = preprocess_img(image, label)  # use singular 'label'

print(f"Image before preprocessing:\n{image[:2]}..., \nShape: {image.shape}, \nDatatype: {image.dtype}\n")
print(f"Image after preprocessing:\n{preprocessed_img[:2]}..., \nShape: {preprocessed_img.shape}, \nDatatype: {preprocessed_img.dtype}")
```

```python
train_data = train_data.map(map_func=preprocess_img,num_parallel_calls = tf.data.AUTOTUNE)
#shuffle

train_data = train_data.shuffle(buffer_size=1000).batch(batch_size=32).prefetch(buffer_size=tf.data.AUTOTUNE)

# map preprocessing fuction to test data

test_data = (
    test_data
    .map(preprocess_img, num_parallel_calls=tf.data.AUTOTUNE)  # preprocess images
    .batch(32)                                                  # batch
    .cache()                                                    # cache for faster access
    .prefetch(tf.data.AUTOTUNE)                                 # prefetch for performance
)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, Model

# -----------------------------
# Build EfficientNetB0 model
# -----------------------------
IMG_SIZE = 128
num_classes = 101  # number of classes in your dataset

input_shape = (IMG_SIZE, IMG_SIZE, 3)
base_model = tf.keras.applications.EfficientNetB0(
    include_top=False,
    weights=None,  # use 'imagenet' if you want pretrained weights
    input_shape=input_shape
)
base_model.trainable = True

inputs = layers.Input(shape=input_shape)
x = base_model(inputs)
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dense(num_classes)(x)
outputs = layers.Activation("softmax", dtype=tf.float32)(x)

model = Model(inputs, outputs, name="EfficientNetB0_FromScratch")

# -----------------------------
# Compile the model
# -----------------------------
model.compile(
    optimizer=tf.keras.optimizers.Adam(),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

#  Model is compiled and ready
model.summary()
```

```python
import tensorflow as tf
import datetime

def create_tensor_callback(dir_name, experiment_name):
  """
  Creates a TensorBoard callback with a specific log directory.
  """
  log_dir = dir_name + "/" + experiment_name + "/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
  tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir)
  print(f"Saving TensorBoard log files to: {log_dir}")
  return tensorboard_callback
```
